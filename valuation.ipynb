```python
pricedf=pricedf[['unique_id', 'price_paid', 'deed_date', 'postcode','saon', 'paon', 'street']]
pricedf.head()
```

```python
pricedf['deed_date']= pd.to_datetime(pricedf['deed_date'])
pricedf['price_paid']= pricedf['price_paid'].astype(int)
pricedf['saon']=pricedf['saon'].astype(str)
pricedf.dtypes
```

```python
pricedf["address"]=pricedf["saon"]+", "+pricedf["paon"]+", "+pricedf["street"]
pricedf.head()
```

```python
from datetime import date
pd.to_datetime(date.today())
```

```python
groupedbydate=pricedf.groupby(['address'])['deed_date'].apply(','.join).reset_index()
groupedbyprice=pricedf.groupby(['address'])['price_paid'].apply(','.join).reset_index()
finaldf=pd.merge(groupedbydate,groupedbyprice,on="address")
uniqueadds=pricedf["address"].value_counts().keys()
finaldf=finaldf.set_index("address")
finaldf.head()
for flat in finaldf.index:
    dates=finaldf.loc[flat,"deed_date"].split(",")
    dates=[pd.to_datetime(date) for date in dates]
    prices=finaldf.loc[flat,"price_paid"].split(",")
    prices=[int(price) for price in prices]
    mindate=min(dates)
    maxdate=max(dates)
    daterangeinyears=((maxdate-mindate).total_seconds())/(60*60*24*365)
    #print(maxdate,mindate,daterangeinyears)
    minprice=min(prices)
    maxprice=max(prices)
    lastprice=prices[0]
    pricerangeinpounds=maxprice-minprice
    priceincreaseperyear=pricerangeinpounds
    #print(maxprice,minprice,pricerangeinpounds)
    finaldf.loc[flat,"date range"]=daterangeinyears
    finaldf.loc[flat,"price range"]=pricerangeinpounds
    periodsincelastsale=(pd.to_datetime(date.today())-maxdate).total_seconds()/(60*60*24*365)
    finaldf.loc[flat,"years since last sale"]=periodsincelastsale
    finaldf.loc[flat,"last price"]=int(lastprice)
    try:
        finaldf["yearly increase"]=finaldf["price range"]/finaldf["date range"]
        finaldf["price now"]=(finaldf["years since last sale"]*finaldf["yearly increase"])+lastprice
    except:
        finaldf["yearly increase"]=0
        finaldf["price now"]=0
finaldf['yearly increase'] = finaldf['yearly increase'].fillna((finaldf["yearly increase"]).mean())
finaldf['price now'] = finaldf['price now'].fillna((finaldf["years since last sale"]*finaldf["yearly increase"])+lastprice)
finaldf["meanincreaseinblock"]=finaldf["yearly increase"].mean()
finaldf["price now (all block projection)"]=(finaldf["years since last sale"]*finaldf["meanincreaseinblock"])+finaldf["last price"]
finaldf["price now (this flat projection)"]=(finaldf["years since last sale"]*finaldf["yearly increase"])+finaldf["last price"]
finaldf
```
